```python
import cv2
import os
import skvideo.io
import math
 
testSet = open("C://Users/gosha/OneDrive/Documents/Summer Research/ucfTrainTestlist/testlist03.txt","r")

def extractFrames(pathIn, pathOut): 
    frameCount = 0
    checkpoint = 0
    framesSaved = 0
    
    cap = cv2.VideoCapture(pathIn)
    
    #############

    numFrames = 0
    while(cap.isOpened()):
      ret, frame = cap.read()
      if(ret == True):
          numFrames+=1
      elif(ret == False):
          break
    frameFreq = numFrames / 20        
    cap.release        
            
    cap = cv2.VideoCapture(pathIn)        
    ##############
    
    while (cap.isOpened()):
 
        ret, frame = cap.read()
        if(ret == True):
            if(frameCount == checkpoint):
              framesSaved += 1
              checkpoint = int(math.floor(frameFreq * framesSaved))##########
              frame = frame[7:234,46 :273]
              cv2.imwrite(os.path.join(pathOut, "Frame{:d}.jpg".format(frameCount)), frame)  # save frame as JPEG file
          
            frameCount += 1
        elif(ret == False):
            break
        
    cap.release()
    

inDirectory = 'C://Users/gosha/OneDrive/Documents/Summer Research/UCF-101/'
outDirectory = 'C://Users/gosha/OneDrive/Documents/Summer Research/TestFrames3.2'    
    
for x in testSet:
  fileLoc = testSet.readline().rstrip('\n')

  if(fileLoc.index('/') > 0):
    splitIndex = fileLoc.index('/')
  else:
    break
 
  foldername = fileLoc[:splitIndex]
  filename = fileLoc[splitIndex+1::]  
  
  if(os.path.exists(outDirectory + '/' + foldername) != True):
     os.mkdir(outDirectory + '/' + foldername)
     
  os.mkdir(outDirectory + '/' + foldername + '/' + filename)
  extractFrames(inDirectory + '/' + fileLoc, outDirectory + '/' + str(fileLoc))
            
        
```
